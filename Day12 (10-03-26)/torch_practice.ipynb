# Torch ops

### torch.add() :

```python
a = torch.ones((2,2))
b= torch.add(a,5)
torch.add(a,1,alpha=5.0,out=a)
print(b)
print(b.dtype)
print(a)
print(a.dtype)
```

### torch.arange() :

```python
print(torch.arange(5))
print(torch.arange(5).dtype)
print(torch.arange(0,1,0.1))
print(torch.arange(0,1,0.1).dtype)
print(torch.arange(1,0,-0.1,dtype=torch.float64))
print(torch.arange(1,0,-0.1,dtype=torch.float64).dtype)
```

```python
x= torch.arange(10.0,requires_grad=True)
y = (x**2).sum()
print(y)
y.backward()
print("derivative of x:", x.grad)
```

### torch.reshape()

```python
a=torch.arange(10)
print(torch.reshape(a,(2,5)))
print(torch.reshape(a,(-1,2)))
a.reshape(-1,)
```

### torch.gather()

```python
a=torch.arange(1,10).reshape(3,3)
idx=torch.tensor([[0,1],[0,1]])
print(a)
torch.gather(a,dim=1,index=idx)
```

### torch.transpose()

```python
a=torch.arange(9).reshape(3,3)
print(a)
print(torch.transpose(a,0,1))
```

### torch.select()

```python
a=torch.arange(6).reshape(2,3)
print(a)
print(a.select(1,0))
print(a.select(1,1))
print(a.select(1,2))
```

### torch.select() :

```python
a= torch.arange(9).reshape(3,3).float()
print(a)
idx=torch.tensor([[2,0,2],[1,1,0],[0,2,1]])
torch.zeros(4,4).scatter_(0,idx,a)
```

### torch.normal() :

```python
print(torch.normal(mean=0,std=0,size=(1,50)))
print(torch.mean(torch.normal(mean=0,std=0,size=(1,50))))
print(torch.normal(mean=0,std=1,size=(1,50)))
print(torch.mean(torch.normal(mean=0,std=1,size=(1,50))))
```
